### EDA and Feature Engineering with Flight Price Prediction Dataset

```python
import pandas as pd
```

```python
df = pd.read_excel('flight_price.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
```

```python
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_hour'] = df['Arrival_Time'].str.split(":").str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(":").str[1]
```

```python
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_min'] = df['Arrival_min'].astype(int)
```

```python
df.head()
```

```python
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df['Departure_hour'] = df['Dep_Time'].str.split(":").str[0]
df['Departure_min'] = df['Dep_Time'].str.split(":").str[1]
```

```python
df.head()
```

```python
df['Departure_hour'] = df['Departure_hour'].astype(int)
df['Departure_min'] = df['Departure_min'].astype(int)
```

```python
df.drop('Dep_Time', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].isnull()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
import numpy as np
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop': 0, '2 stops': 2, '1 stop': 1, '3 stops': 3, np.nan: 1, '4 stops': 4})
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.head(
)
```

```python
df.drop('Route', axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df.info()
```

```python
df['Duration_hour'] = df['Duration'].str.split(" ").str[0].str.split('h').str[0]
df['Duration_min'] = df['Duration'].str.split(" ").str[1].str.split('m').str[0]
```

```python
df.head()
```

```python
df[df['Duration_min'].isnull()]
```

```python
df['Duration_min'] = df['Duration_min'].fillna(0)
```

```python
df[df['Duration_min'].isnull()]
```

```python
df.info()
```

```python
df['Duration_min'].unique()
```

```python
df['Duration_hour'].unique()
```

```python
df['Duration_hour'] = df['Duration_hour'].astype(int)
df['Duration_min'] = df['Duration_min'].astype(int)
```

```python
df.drop('Duration', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
encoder.fit_transform(df[['Airline', 'Source', 'Destination', 'Additional_Info']]).toarray()
```

```python
df_encoded = pd.DataFrame(encoder.fit_transform(df[['Airline', 'Source', 'Destination', 'Additional_Info']]).toarray(), columns = encoder.get_feature_names_out())
```

```python
df.head()
```

```python
df = pd.concat([df, df_encoded], axis=1)
```

```python
df.drop(['Airline', 'Source', 'Destination', 'Additional_Info'], axis=1, inplace=True)
```
